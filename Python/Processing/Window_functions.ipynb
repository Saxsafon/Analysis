# Оконные функции

[Pandas doc: Windowing operations](https://pandas.pydata.org/docs/user_guide/window.html)

Найти максимальное значение для каждой группы.

```python
data.head()
```

**groupby + transform**

```python
data['max_value_1'] = data.groupby('group')['value'].transform('max')
```

**groupby + merge**

```python
data = data.merge(
    data.groupby('group')['value'].apply(max).rename('max_value_2').reset_index(),
    on='group', how='left'
)
```

```python
data = data.sort_values('group', ascending=True)
data
```
