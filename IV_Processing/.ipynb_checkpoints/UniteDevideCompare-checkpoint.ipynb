```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'col1': [1,2,3],
    'col2': ['A','B','C'],
    'col3': ['a','b','c']
})

df2 = pd.DataFrame({
    'col1': [4,5,6],
    'col2': ['D','E','F'],
    'col3': ['d','e','f']
})
```

```python
pd.concat([df1, df2], axis=0, ignore_index=False)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=False)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=True)
```

```python
df1.append(df2)
```
