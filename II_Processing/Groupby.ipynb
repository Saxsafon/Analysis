# Группировки

## Методы диспетчеризации

Все методы, не реализованные явным образом объектом GroupBy, будут передаваться далее и выполняться для групп, вне зависимости от того, являются ли они объектами Series или DataFrame. 

Например, можно использовать метод describe() объекта DataFrame для вычисления набора сводных показателей, описывающих каждую группу в данных 

```python
data.groupby('group')['value'].describe()
```

## Метод apply

### Ввести колонку с комбинацией групп в течении дня

```python
add = (
    data
        .groupby('date')
        .apply(lambda day: '_'.join(sorted(day['group'].unique())))
        .reset_index()
        .rename(columns={0: 'group_combination'})

)

add
```

```python
data = data.merge(add, on='date')
data.head()
```

### Посчитать несколько новых колонок

```python
def f(day):
    d = {
        'sub_group_combination': '_'.join(sorted(day['sub_group'].unique())),
        'sub_group_nunique': day['sub_group'].nunique(),
        'val_id_nunique': day['val_id'].nunique(),
    }
    
    return pd.Series(d)
    
add = data.groupby('date').apply(f).reset_index()
add
```

```python
data = data.merge(add, on='date')
data.head()
```

## Фильтрация

Операции фильтрации дает возможность опускать данные в зависимости от свойств группы. Например нам может подадобиться оставить в результате все группы, в которых стандартное отклонение превышает какое-либо критическое значение.

```python
def filter_func(x):
    return x.value.std() > 24
```

```python
data.groupby('group')['value'].std()
```

```python
data.groupby('group').filter(filter_func)[['group', 'value']]
```

Функция filter() возвращает будево значение, определяющее, прошла ли группа фильтрацию. В данном случае, только у группы C стандартное отклонение превышает 24

## Кастомные функции

### Просумировать уникальные значения для каждого уровня группировки

```python
def groupcount(df, cols, char, prefix="_count"):
    grouplist = []
    
    for col in cols:
        grouplist.append(col)
        add = df.groupby([*grouplist])[char].nunique().reset_index()
        
        old, new = add.iloc[:,-1].name, f"{col}{prefix}"
        add = add.rename(columns={old: new})
        
        add = add[[*grouplist, new]]
        df = pd.merge(df, add, on=[*grouplist])
        grouplist.append(new)
        
    gc = df.groupby([*grouplist[:-1]])[char].nunique().to_frame().rename(columns={char: grouplist[-1]})
    return gc
```

```python
groupcount(data, cols=['date', 'hour', 'group', 'sub_group'], char='val_id', prefix="_count")
```
